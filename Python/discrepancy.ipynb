```python
import numpy as np
import matplotlib.pyplot as plt
```

## Uniform random samples
Used as baseline.

```python
def uniform_random_samples(n: int):
    x = np.random.uniform(0, 1, size=n)
    y = np.random.uniform(0, 1, size=n)
    return (x, y)
urs_x, urs_y = uniform_random_samples(256)
fig = plt.figure(figsize=(6, 6), dpi=80)
plt.scatter(urs_x, urs_y)
plt.show()
```

## Van der Corput sequence

1D sequence.

**Radical inversion**: $\Phi_{b, C}(i)$

$$\begin{array}{l}
i=\sum_{l=0}^{M-1} a_{l}(i) b^{l} \\
\Phi_{b, C}(i)=\left(b^{-1} \ldots b^{-M}\right)\left[C\left(a_{0}(i) \ldots a_{M-1}(i)\right)^{T}\right]
\end{array}$$

And Van der Corput sequence is defined as: 

$$
\Phi_{b}(i)=\left(b^{-1} \ldots b^{-M}\right)\left(a_{0}(i) \ldots a_{M-1}(i)\right)^{T}=\sum_{l=0}^{M-1} a_{l}(i) b^{-l-1}
$$

In which, $C$ is the identity matrix $I$.

```python
def vandercorput_random_sample(b: int, n: int, start: int = 1):
    x = np.zeros((n, 1), dtype=np.float64)
    for i in range(start, n + start):
        bi = 1.0
        r = 0.0
        j = i
        while j:
            bi /= b
            s = j % b
            r += bi * s
            j = j // b
        x[i - start]  = r
    return x
vrs_x = vandercorput_random_sample(2, 64)
fig = plt.figure(figsize=(6, 6), dpi=80)
plt.scatter(vrs_x, np.random.uniform(0.4, 0.6, 64), s=5.0)
plt.ylim((0.0, 1.0))
plt.show()
```
